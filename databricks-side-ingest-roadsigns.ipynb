```python
%fs
mkdirs dbfs:/FileStore/GTSRB
```

```python
%sh
# COPIES TO DBFS

for X in $(find GTSRB |grep png); do
  label=$(echo $X |cut -d / -f 4)
  mkdir -p /dbfs/FileStore/GTSRB/label=$label
  cp $X /dbfs/FileStore/GTSRB/label=$label/
  echo "done $X"
done
```

```python
df = spark.read \
    .format("image") \
    .load("dbfs:/FileStore/GTSRB")
```

```python
df.count()
```

```python
df.write.format("delta").mode("overwrite").save("/delta/gtsrb")
```

```python
spark.sql("CREATE TABLE gtsrb USING DELTA LOCATION '/delta/gtsrb/'")
```
